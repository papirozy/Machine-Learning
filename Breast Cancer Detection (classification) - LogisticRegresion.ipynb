```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split # to split features into Training and Test
from sklearn.linear_model import LogisticRegression # importing the Logistic Regression Algorithm
from sklearn.metrics import accuracy_score # this is for model evaluation; to know how many correct predictions were made
```

```python
# Our dataset comes with a diagnosis column which holds all predictions
# The diagnosis column holds 2 values of outputs/Classififctions (M - malignant; B - benign)
```

```python
# Collect Data
cancer_df = pd.read_csv('brca.csv')
cancer_df.head()
```

```python
# NOTICE: The 1st column holdssome unwanted data which we would delete.
cancer_df = cancer_df.drop(columns=['Unnamed: 0'], axis=1)
cancer_df
```

```python
# Let's do some Data Analysis:
# Number of ROWS & COLUMNS
cancer_df.shape       #   569 Rows | 31 Columns (30 columns for inputs; 1 column for Output/prediction)
```

```python
# Info on Columns - (Data types & Null values)
cancer_df.info()
```

```python
# check for missing values (which already can be viewed from query above)
cancer_df.isnull().sum()
```

```python
# more stats about the dataset
cancer_df.describe()
```

```python
# Checking the OUTPUT column for available classifications
cancer_df['y'].value_counts()  # 357 Benign cases | 212 Malignant cases
```

```python
# group the dataset based on the classes above
cancer_df.groupby('y').mean()
```

```python
# Out Output column 'y' housing PREDICTIONS has values "B" & "M"
# String values are pretty alien to machine language(computations)
# Hence the need to convert these strings to numberic values (encoding - called labelEncoding)
# We are to convert values "M" & "B" into numeric values
# import te encoding library
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the LabelEncoder function into a variable for work to commence
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(cancer_df.y)   # converts labels in the "y" column, storing them in this variable
```

```python
# Create a new column in our dataset to hold the converted labels above
cancer_df['diagnosis'] = labels
```

```python
cancer_df['diagnosis'].value_counts()
```

```python
# Drop column 'y' since it would'nt be needed anymore
cancer_df.drop(columns=['y'], axis=1)
```

```python
# View the dataset for evidence 
cancer_df.head()
```

```python
cancer_df.groupby('diagnosis').mean(numeric_only=True)
```

```python
pd.__version__
```

```python
# Let's split our Dataset in preparation for Model training
X = cancer_df.drop(columns=['diagnosis','y'])
Y = cancer_df['diagnosis']
```

```python
print(Y)
```

```python
# Split data into Training & Test
# Training data wil be used to train the model
# Test data will be used for evaluation of our Model's performance.
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Commence Model Training
logRegModel = LogisticRegression()
```

```python
# Fit the Dataset
logRegModel.fit(X_train, Y_train)
```

```python
# Evaluate our Model
# For classification we use a metric called accuracy_score
# Find the Accuracy Score on the Training Data
# WE will submit the X_train data to our model to see how accurately it will predict corresponding Y_train values
training_prediction = logRegModel.predict(X_train)
```

```python
# the above training_prediction values will be compared with Y_train values for prediction accuracy
training_accuracy_score = accuracy_score(Y_train, training_prediction)
print('Accuracy Score (training): ', training_accuracy_score)
```

```python
# Accuracy is 95%
```

```python
# Find the TEST Data accuracy for your Model
test_prediction = logRegModel.predict(X_test)
test_accuracy_score = accuracy_score(Y_test, test_prediction)
print('Accuracy score (TEST): ', test_accuracy_score)
```

```python
# Accuracy is 94%
```
